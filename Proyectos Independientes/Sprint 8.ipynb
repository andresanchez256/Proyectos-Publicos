# Paso 5: Prueba de hipótesis (Python)

# Cargar y preparar datos de viajes Loop -> O'Hare

```python
print("\n" + "="*60)
print("PRUEBA DE HIPÓTESIS")
print("="*60)
```

# Verificar la distribución de datos por condición climática

```python
print("\n📊 Distribución de viajes por clima:")
print(df_trips['weather_conditions'].value_counts())
```

# Separar los datos en dos grupos

```python
bad_weather = df_trips[df_trips['weather_conditions'] == 'Bad']['duration_seconds']
good_weather = df_trips[df_trips['weather_conditions'] == 'Good']['duration_seconds']
```

# Estadísticas descriptivas por grupo

```python
print("\n📈 Estadísticas descriptivas:")
print(f"Clima MALO (rain/storm):")
print(f"  - Número de viajes: {len(bad_weather)}")
print(f"  - Duración promedio: {bad_weather.mean():.2f} segundos ({bad_weather.mean()/60:.2f} minutos)")
print(f"  - Desviación estándar: {bad_weather.std():.2f} segundos")

print(f"\nClima BUENO:" )
print(f"  - Número de viajes: {len(good_weather)}")
print(f"  - Duración promedio: {good_weather.mean():.2f} segundos ({good_weather.mean()/60:.2f} minutos)")
print(f"  - Desviación estándar: {good_weather.std():.2f} segundos")
```

# Visualización de la distribución

```python
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# Histogramas

axes[0].hist(good_weather, bins=30, alpha=0.7, label='Buen clima', color='lightgreen', edgecolor='black')
axes[0].hist(bad_weather, bins=30, alpha=0.7, label='Mal clima', color='lightcoral', edgecolor='black')
axes[0].set_xlabel('Duración del viaje (segundos)')
axes[0].set_ylabel('Frecuencia')
axes[0].set_title('Distribución de duraciones por condición climática')
axes[0].legend()

# Boxplot

bp = axes[1].boxplot([good_weather, bad_weather], 
                      labels=['Buen clima', 'Mal clima'],
                      patch_artist=True,
                      showmeans=True,
                      meanprops={'marker': 'D', 'markerfacecolor': 'red', 'markersize': 8})
bp['boxes'][0].set_facecolor('lightgreen')
bp['boxes'][1].set_facecolor('lightcoral')
axes[1].set_ylabel('Duración del viaje (segundos)')
axes[1].set_title('Comparación de duraciones (boxplot)')

plt.tight_layout()
plt.show()
```

# PRUEBA DE HIPÓTESIS

```python
"""
FORMULACIÓN DE HIPÓTESIS:

Hipótesis nula (H0): 
    La duración promedio de los viajes desde Loop hasta O'Hare los sábados 
    NO cambia cuando hay mal clima (lluvia/tormenta). Es decir, la duración 
    media con mal clima es igual a la duración media con buen clima.
    
    H0: μ_bad = μ_good

Hipótesis alternativa (H1): 
    La duración promedio de los viajes desde Loop hasta O'Hare los sábados 
    SÍ cambia cuando hay mal clima. Las duraciones son diferentes.
    
    H1: μ_bad ≠ μ_good
"""
```

# SELECCIÓN DEL CRITERIO DE PRUEBA

```python
print("\n" + "="*60)
print("SELECCIÓN DEL CRITERIO DE PRUEBA")
print("="*60)

print("""
CRITERIO: Prueba t de Student para dos muestras independientes

RAZONES DE LA SELECCIÓN:
1. Las dos muestras (clima bueno y malo) son independientes
2. Estamos comparando las medias de dos grupos
3. El tamaño de muestra es grande (>50 en ambos grupos)
4. Por el Teorema Central del Límite, la distribución de medias es normal

PARÁMETROS:
- Alfa (nivel de significación) = 0.05 (5%)
- Prueba de dos colas (porque queremos detectar cualquier diferencia)
""")
```

# Realizar prueba t

```python
t_statistic, p_value = stats.ttest_ind(bad_weather, good_weather, equal_var=False)

print(f"\n📊 RESULTADOS DE LA PRUEBA:")
print(f"Estadístico t: {t_statistic:.4f}")
print(f"Valor p: {p_value:.6f}")
print(f"Nivel de significancia (α): 0.05")
```

# CONCLUSIÓN DE LA HIPÓTESIS

```python
print("\n" + "="*60)
print("CONCLUSIÓN DE LA PRUEBA DE HIPÓTESIS")
print("="*60)

alpha = 0.05
if p_value < alpha:
    print(f"✅ RECHAZAMOS la hipótesis nula (p-valor = {p_value:.6f} < {alpha})")
    print("\nConclusión: Los sábados con mal clima (lluvia/tormenta), la duración")
    print("de los viajes desde Loop hasta O'Hare ES SIGNIFICATIVAMENTE DIFERENTE")
    print("comparado con los sábados de buen clima.")
    
    # Calcular la diferencia práctica
    diff = abs(bad_weather.mean() - good_weather.mean())
    diff_minutes = diff / 60
    print(f"\nDiferencia práctica: {diff:.0f} segundos ({diff_minutes:.1f} minutos)")
    
    if bad_weather.mean() > good_weather.mean():
        print("Los viajes duran MÁS tiempo cuando llueve/hay tormenta")
    else:
        print("Los viajes duran MENOS tiempo cuando llueve/hay tormenta")
else:
    print(f"❌ NO podemos rechazar la hipótesis nula (p-valor = {p_value:.6f} ≥ {alpha})")
    print("\nConclusión: No hay evidencia estadística suficiente para afirmar que")
    print("las condiciones climáticas adversas afecten la duración de los viajes")
    print("desde Loop hasta O'Hare los sábados.")
```

 # Cálculo adicional: intervalo de confianza

```python
print("\n" + "="*60)
print("INTERVALOS DE CONFIANZA (95%)")
print("="*60)

from scipy import stats as st
```

# Calcular intervalos de confianza

```python
bad_ci = st.t.interval(0.95, len(bad_weather)-1, loc=bad_weather.mean(), scale=st.sem(bad_weather))
good_ci = st.t.interval(0.95, len(good_weather)-1, loc=good_weather.mean(), scale=st.sem(good_weather))

print(f"Clima MALO: IC 95% = ({bad_ci[0]:.0f}, {bad_ci[1]:.0f}) segundos")
print(f"            = ({bad_ci[0]/60:.1f}, {bad_ci[1]/60:.1f}) minutos")
print(f"\nClima BUENO: IC 95% = ({good_ci[0]:.0f}, {good_ci[1]:.0f}) segundos")
print(f"             = ({good_ci[0]/60:.1f}, {good_ci[1]/60:.1f}) minutos")

if p_value < alpha:
    print("\n⚠️ Los intervalos de confianza NO se superponen, confirmando diferencia significativa")
else:
    print("\n✅ Los intervalos de confianza se superponen, consistente con H0")
```
