```python
#Bases de datos
a80     = pd.read_table("anime80s.csv",header=0)
a90     = pd.read_table("anime90s.csv",header=0)
a00     = pd.read_table("anime00s.csv",header=0)
a10     = pd.read_table("anime10s.csv",header=0)
animes  = pd.concat([a80,a90,a00,a10])

animes.info()
animes
```

```python
ratings = pd.read_table("animeratings.csv",header=0)
ratings.info()
ratings
```

```python
ratings2=(ratings
         .groupby(["anime_id"])
         .agg(veces=("anime_id","count"),rtgp=("rating","mean")).reset_index()
         .query("veces>=10000").reset_index(drop=True) 
         )
ratings2.info()
ratings2
```

```python
animes2 = (ratings2
          .merge(animes,left_on="anime_id",right_on="MAL_ID",how="left")
          .query("Genres.str.contains('(^| )Action').values & Type=='TV'")
          .loc[:,["anime_id","rtgp","Name","English name","Episodes","Studios","Source","Premiered"]].reset_index(drop=True)  
          )
animes2.info()
animes2
```
