```python
import scipy.stats as stats
import numpy as np

# Ejemplo 1: ¿Cuántos fracasos hasta obtener 5 éxitos con p=0.4?
# con probabilidad acumulada del 90%
n_exitos = 8
p_exito = 0.6
prob = 0.99

fracasos = stats.nbinom.ppf(prob, n_exitos, p_exito)
print(f"Para obtener ", n_exitos, " éxitos con p=",p_exito,", en el ",100*prob,"% de los casos")
print(f"se necesitan {fracasos} fracasos o menos")
print(f"(Total de ensayos: {fracasos + n_exitos})")
# Salida: 17 fracasos (22 ensayos totales)

# Ejemplo 2: Múltiples probabilidades a la vez
probs = np.array([0.25, 0.5, 0.75, 0.95])
cuantiles = stats.nbinom.ppf(probs, n=n_exitos , p=p_exito)
print(f"\nCuantiles para diferentes probabilidades:")
print(f"25%: {cuantiles[0]} fracasos")
print(f"50%: {cuantiles[1]} fracasos")
print(f"75%: {cuantiles[2]} fracasos")
print(f"95%: {cuantiles[3]} fracasos")
```
